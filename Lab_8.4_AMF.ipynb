## PART 1

### 1. Establishing the Null and Alternate Hypothesis

H0: *there is no difference in the mean etching rate for different levels of power*

H1: *there is a significant difference in the mean etching rate for different levels of power (in at least a level of power)*

### 2. What is the significance level

Significance Level chosen: α=0.05

* It is the probability of rejecting the null hypothesis when it is true. 
* If we set the significance level at 0.05 we say there is a 5% probability of rejecting H0 when it is true (saying that there is a difference in means when there is actually no difference)

### 3. What are the degrees of freedom of model, error terms, and total DoF

**Degrees of Freedom**. In this case the degrees of freedom are:

- **DoF Model**: 3-1 = **2**

- **DoF Error**: 15-3 = **12**

- **Total DoF**: 15-1 = **14**

## PART 2

use Python to conduct ANOVA

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
```

```python
data = pd.read_excel('anova_lab_data.xlsx', sheet_name='data_collected')
data
```

```python
data.rename(columns={'Etching Rate':'Etching_Rate'}, inplace=True)
#let's rename the variables so that we can make the ANOVA work
```

```python
data.rename(columns={'Power ':'Power'}, inplace=True)
#let's rename the variables so that we can make the ANOVA work
```

```python
##ANOVA:
model = ols('Etching_Rate ~ C(Power)',data=data).fit()
table = sm.stats.anova_lm(model)
print(table)
```

What conclusions can you draw from the experiment and why?

The ANOVA results show a **p-value from the F-test under 0.05**. This means **we can reject the H0**. This is, **there are significant differences amongst the mean etching rate for at least a level of power**. 

Also, by loooking at the F-statistic, its value is 36.88. By looking at the F-statistic table the **critical value** for the given significance level (0.05) and the degrees of freedom (2 and 12), is **3.89**.

Since F0 (test statistic): 36.88 > 3.89 (F_critical), **we also reject the null hypothesis**
